Load NFL combine stats to use as sample data

```python
dfs = pd.read_html('http://nflcombineresults.com/nflcombinedata.php?year=all&pos=&college=')
```

```python
len(dfs)
```

```python
df = dfs[0]
```

Add column names

```python
df.columns = ['Year', 'Name', 'College', 'POS', 'Height (in)', 'Weight (lbs)', 'Wonderlic', '40 Yard', 'Bench Press', 'Vert Leap', 'Broad Jump', 'Shuttle', '3Cone']
```

Remove asterisks from 40 Yard times

```python
df['40 Yard'] = df['40 Yard'].apply(lambda x: float(str(x).replace('*', '')))
```

Draw a bean plot of bench press, grouping players by position and sorting by the mean bench press for each group

```python
plots.draw_bplot(df, 'POS', 'Bench Press', sort_by = 'Bench Press', sort_func = np.mean)
```
